## Test request for the country France

```python
#Make a GET request
r = requests.get("https://www.henleypassportindex.com/fetch?url=passports%2FFR%2Fcountries")
```

```python
#See what response code the server sent back
r.status_code
```

```python
#Check the responseâ€™s Content Type
r.headers.get("content-type", "unknown")
```

```python
#Access the full response as text
print(r.text)
```

```python
#Using Regular Expressions to look for HTML patterns
# 1 = French people can enter the country visa-fee
# 0 = French people can NOT enter the country visa-fee
re.findall(r'(?i)(?:\[?\{"name":"[\w+\s]*","flags":\[)?\[?\{"id":\d+,"name":"([\w+\(\)\s\-\.\'\;]*)","code":"(\w+)","longitude":null,"latitude":null,"show_map":0,"created_at":"\d+-\d+-\d+ \d+:\d+:\d+","updated_at":"\d+-\d+-\d+ \d+:\d+:\d+","deleted_at":null,"region_id":\d+,"pivot":\{"passport_id":\d+,"country_id":\d+,"visa_access_id":\d+,"is_visa_free":(\d+)},"region":\{"id":\d+,"name":"[\w+\s]*"\,"deleted_at":null,"created_at":"\d+-\d+-\d+ \d+:\d+:\d+"\,"updated_at":"\d+-\d+-\d+ \d+:\d+:\d+"\}\}(?:\]\})?\,?\]?\s*', r.text)
```
